```python
hotel_bookings.rename({'adults': 'num_adults', 'children': 'num_children', 'babies': 'num_babies'}, axis=1, inplace=True)# Renaming columns for clarity
hotel_bookings.head()
```

```python
hotel_bookings.isnull().sum()*100 / len(hotel_bookings)
```

```python
hotel_bookings['agent'].unique()
```

```python
hotel_bookings['agent'].unique()# Displaying unique agent values
hotel_bookings[hotel_bookings['agent'] == 5]# Checking for rows where agent is 5

hotel_bookings['agent'].fillna(-1, inplace=True)# Filling missing agent values with -1

hotel_bookings[hotel_bookings['num_children'].isna()]#  Checking for rows with missing num_children values

hotel_bookings[hotel_bookings['country'].isna()][['hotel', 'is_canceled', 'country']]# Checking for rows with missing country values

hotel_bookings['agent'].fillna('Unknown', inplace=True)# Filling missing agent values with 'Unknown'

hotel_bookings.dropna(subset=['num_children'], inplace=True )# Dropping rows with missing num_children values
```
